```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'heart_disease_for_tableau_final.csv'
heart_disease_data = pd.read_csv(file_path)

# Visualization 1: Cholesterol Levels by Heart Disease Presence
plt.figure(figsize=(8, 6))
sns.boxplot(data=heart_disease_data, x='Target_readable', y='chol', hue='Target_readable', palette='Set2')
plt.title('Cholesterol Levels by Heart Disease Presence', fontsize=14)
plt.xlabel('Heart Disease Presence', fontsize=12)
plt.ylabel('Cholesterol Level', fontsize=12)
plt.legend(title='Heart Disease Presence')
plt.show()


# Visualization 2: Scatter Plot of Cholesterol vs. Age by Heart Disease Presence
plt.figure(figsize=(8, 6))
sns.scatterplot(data=heart_disease_data, x='age', y='chol', hue='Target_readable', palette='Set1')
plt.title('Cholesterol vs. Age by Heart Disease Presence', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Cholesterol Level', fontsize=12)
plt.legend(title='Heart Disease Presence')
plt.show()

# Statistical Summary of Cholesterol by Heart Disease Presence
cholesterol_summary = heart_disease_data.groupby('Target_readable')['chol'].describe()
print(cholesterol_summary)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'heart_disease_for_tableau_final.csv'  
heart_disease_data = pd.read_csv(file_path)

# Visualization 1: Maximum Heart Rate vs Age by Heart Disease Presence
plt.figure(figsize=(8, 6))
sns.scatterplot(data=heart_disease_data, x='age', y='max_heart_rate', hue='Target_readable', palette='coolwarm')
plt.title('Maximum Heart Rate vs Age by Heart Disease Presence', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Maximum Heart Rate', fontsize=12)
plt.legend(title='Heart Disease Presence')
plt.show()

# Visualization 2: Distribution of Chest Pain Type by Heart Disease Presence
plt.figure(figsize=(8, 6))
sns.countplot(data=heart_disease_data, x='cp_readable', hue='Target_readable', palette='Set2')
plt.title('Chest Pain Type Distribution by Heart Disease Presence', fontsize=14)
plt.xlabel('Chest Pain Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Heart Disease Presence')
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'heart_disease_for_tableau_final.csv'  
dataset = pd.read_csv(file_path)

# Step 0: Inspecting the dataset's structure
print("Dataset Shape:", dataset.shape)
print("\nDataset Info:")
print(dataset.info())




# Step 1: Histogram for Age Distribution
plt.figure(figsize=(8, 4))
sns.histplot(dataset['age'], kde=True, bins=10, color='blue')
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()



# Step 2: Correlation Heatmap
numerical_columns = dataset.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = dataset[numerical_columns].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

# Step 3: Pairplot of Selected Features Colored by Target
selected_columns = ['age', 'chol', 'trestbps', 'max_heart_rate', 'oldpeak', 'Target']
sns.pairplot(dataset[selected_columns], hue='Target', diag_kind='kde')
plt.suptitle("Relationships Between Selected Features", y=1.02)
plt.show()

# Boxplot for Oldpeak and Target
plt.figure(figsize=(8, 4))
sns.boxplot(x='Target', y='oldpeak', data=dataset, hue='Target', dodge=False)  
plt.title("Relationship Between Oldpeak and Target")
plt.xlabel("Target (0 = Absence, 1 = Presence)")
plt.ylabel("Oldpeak")
plt.legend([], [], frameon=False)  
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load the dataset
file_path = 'heart_disease_for_tableau_final.csv'
heart_disease_data = pd.read_csv(file_path)

# Preprocess data for visualization
# For the Tree Map
tree_map_data = heart_disease_data.groupby('age_group')['chol'].sum().reset_index()



# Tree Map Plot
plt.figure(figsize=(10, 6))
sizes = tree_map_data['chol']
labels = tree_map_data['age_group']
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.viridis(np.linspace(0, 1, len(labels))),
)
plt.title("Cholesterol Distribution by Age Group (Tree Map )")
plt.show()

```
